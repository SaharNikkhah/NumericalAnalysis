Prove the following error bound for the Taylor series:

Let $f:\mathbb{R}\to\mathbb{R}$ be $k+1$ times continuously differentiable on the open interval with $f^{(k)}$ continuous on the closed interval from $x_0$ to $x$. Then
$$R_k(x) = \frac{f^{(k+1)}(\xi)}{(k+1)!}(x-x_0)^{k+1}$$
for some $\xi$ between $x$ and $x_0$.

Hint: Mean value theorem

We Can write:
$$R_0(x)=f(x)-T_0(x) $$ 
$$R_0(x)=f(x)-f(x_0) $$ 
Where R is the error f is the function and T is the Taylor expansion. 
f is differentiable on the interval containing (${x_0}$, x), By Mean Value Theorem we can say that there exists a number $\xi$ between $x_0$ and x that:
$$f'(\xi)= \frac {f(x)-f(x_0)}{x-x_0} $$ 
so
$$R_0(x)=f'(\xi)(x-x_0)$$ 
By using the MTV again, and knowing that f is k+1 times differentiable on the interval of (x, $x_0$), the kth R can be written as: 

$$ R_k(x)=\frac {f^{k+1}(\xi)}{(k+1)!}(x-x_0)^{k+1}$$

Recall Neville's algorithm:
\begin{equation*}
\begin{split}
p_{i,i} &= y_i\\[2pt]
p_{i,j} &= \frac{(x-x_i)p_{i+1,j}(x) - (x-x_j)p_{i,j-1}(x)}{x_j-x_i}
\end{split}
\end{equation*}
Write a Python function that implements this algorithm. The function should have three inputs: a list $x_i$, a list $y_i$, and a float $x$. The two lists should be the same length and contain the data while the float is the value we want to evaluate the interpolating polynomial at.



```python
#This program does Nevill's Collumn by Collumn
x=[1,2,3,5]
y=[1,4,9,25]
X=4
T=y

for i in range (len(y)):
    for j in reversed(range(0,i)):
        T[j]= T[j+1]+((T[j+1]-T[j])*((X-x[i])/(x[i]-x[j])))
        #print(i, j, T, T[j])

print(T[0])
```
